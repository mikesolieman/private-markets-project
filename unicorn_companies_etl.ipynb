```python
# Import libraries
import pandas as pd
from ast import literal_eval
import warnings

warnings.simplefilter(action="ignore", category=FutureWarning)
```

```python
# Read CSV file
df = pd.read_csv('/Users/mikesolieman/Documents/Projects/8_Unicorn_Companies/Unicorn_Companies.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
# Drop rows with NaN values
df.dropna(inplace=True)
df.info()
```

```python
# Remove "$" and "B" and cast "Valuation" as float
df['Valuation'] = (df['Valuation']
    .str.replace('$', '')
    .str.replace('B', '000000000')
    .astype(float)
)
```

```python
df.head()
```

```python
# Check for non-numerical values
df['Funding'].value_counts()    
```

```python
# Exclude rows with unknown "Funding" values
df = df[df['Funding'] != 'Unknown']
df.shape
```

```python
# Remove "$" and "B" and cast "Funding" as float
df['Funding'] = (df['Funding']
    .str.replace('$', '')
    .str.replace('B', '000000000')
    .str.replace('M', '000000')
    .astype(float)
)
df.head()
```

```python
df.info()
```

```python
# Export cleaned data as csv to folder
df.to_csv('unicorn_companies_clean_grouped.csv', index=False)
```

```python
# Split investors and convert to list
df['Select Investors'] = df['Select Investors'].str.split(', ')
df.head()
```

```python
df.info()
```

```python
# Explode "Select Investors" to individual list
df = df.explode("Select Investors")
df.head()
```
